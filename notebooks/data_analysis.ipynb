### Building and Training

In this section we will build our models and train them. First, let's create a a dictionary where we will save all the raw data from every trained model.

```python
info = {}
```

```python
def saveToInfo(info, model):
    
    name = type(model.best_estimator_).__name__
    info[name] = {}
    
    # Save the scoring method
    info[name]['scoring'] = model.scoring
    
    # Save best score
    info[name]['best_score'] = model.best_score_
    
    # Save the best param
    info[name]['best_param'] = model.best_params_
    
    # Save the CV_Results
    info[name]['cv_results'] = model.cv_results_
    
    return info
```

```python
def saveEstimator(model):
    
    name = type(model.best_estimator_).__name__
    
    filename = f"{PATH}backend\\models\\oversample\\{name}.pkl"
    outfile = open(filename, 'wb')
    pickle.dump(model.best_estimator_, outfile)
    outfile.close()
```

#### Neural Network

```python
nn_params={
    'epochs':[1,10,25,50],
     'batch_size':[64,128],
     'layers':[2,3,4],
     'units':[256,128,64],
     'activation':['relu'],
     'dropout_rate':[0.2,0.3,0.5],
     'optimizer__lr': [0.05, 0.1]
}
```

```python
def NeuralNetwork(meta, **kwargs):
    
    n_features_in_ = meta["n_features_in_"]
    n_classes_ = meta["n_classes_"]
    
    layers = kwargs.get('layers')
    units = kwargs.get('units')
    activation = kwargs.get('activation')
    dropout_rate = kwargs.get('dropout_rate')
    
    model = Sequential()
    model.add(Dense(units=units, input_dim=n_features_in_, activation = activation))
    for i in range(layers):
        units = int(units/2);
        if units > 1:
            model.add(Dense(units=units, activation=activation))
            model.add(Dropout(dropout_rate))
        else:
            break
    model.add(Dense(1, activation='sigmoid'))
    return model
```

```python
params = {
    'layers':3,
    'units':64,
    'activation':'relu',
    'dropout_rate':0.2,
}

compile_kwargs = {
    'optimizer':'adam',
    'optimizer__lr':0.01,
    'loss':'binary_crossentropy',
    'metric':'accuracy'
}
```

```python
neuralnet = KerasClassifier(model=NeuralNetwork, **params, **compile_kwargs)
```

```python
nn_model = RandomizedSearchCV(neuralnet, nn_params, cv=10, scoring='roc_auc', refit="roc_auc")
```

Now, let's calculate the metrics.

```python
metrics = [
    sklearn.metrics.accuracy_score,
    sklearn.metrics.confusion_matrix,
    sklearn.metrics.precision_score,
    sklearn.metrics.recall_score,
    sklearn.metrics.f1_score,
    sklearn.metrics.roc_curve,
    sklearn.metrics.roc_auc_score
]
```

```python
def evaluate_model(model, X_test, y_test, metrics):
    
    evaluation = {}
    
    y_pred = model.predict(X_test)
    
    for i in range(len(metrics)):
        
        result = metrics[i](y_test,y_pred)
        name = metrics[i].__name__
        evaluation[name] = result
    
    return evaluation
```

```python
metrics_names = [metrics[i].__name__ for i in range(len(metrics))]
```

```python
evals = {}
evals_oversampling = {}

for score_name in metrics_names:
    evals[score_name] = []
    evals_oversampling[score_name] = []
    
    for model, scores in total_eval.items():
        evals[score_name] += [scores[score_name]]
        
    for model, scores in total_eval_oversample.items():
        evals_oversampling[score_name] += [scores[score_name]]
```
